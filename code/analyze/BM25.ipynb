```python
import json

documents = [] # letters
file_path = "/content/feature-matching-1.jsonl"

with open(file_path, 'r') as f:
    for line in f:
        try:
            data = json.loads(line)
            if 'text' in data:
                documents.append(data['text'])
        except json.JSONDecodeError:
            print(f"Skipping invalid JSON line: {line}")

print(f"Loaded {len(documents)} strings from {file_path}")
```

```python
from rank_bm25 import BM25Okapi
import re

# Function to tokenize the text
def tokenize(text):
    text = text.lower()
    text = re.sub(r'\W+', ' ', text) # remove non-alphanumeric characters
    tokens = text.split()
    return tokens

# Tokenize documents
tokenized_documents = [tokenize(doc) for doc in documents]

# Create BM25 object
bm25 = BM25Okapi(tokenized_documents)

# Rank documents for each query
results = {}
for i, query in enumerate(queries):
    tokenized_query = tokenize(query)
    scores = bm25.get_scores(tokenized_query)
    ranked_docs_indices = sorted(range(len(scores)), key=lambda k: scores[k], reverse=True)
    # Store query ID, document ID and score
    ranked_docs = [(i, idx, scores[idx]) for idx in ranked_docs_indices]
    results[f"Query {i}"] = ranked_docs

# Display results for the first few queries as an example
for query_id, ranking in list(results.items())[:47]:
    print(f"\nRanking for {query_id}:")
    for query_id, doc_id, score in ranking[:5]: # Display top 5 documents
        print(f"Score: {score:.4f}, Query ID: {query_id}, Document ID: {doc_id}")
```
